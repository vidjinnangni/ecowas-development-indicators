# Exploration indicateurs économiques CEDEAO

Notebook pédagogique.

**Prérequis** : avoir exécuté `scripts/fetch_data.py`
puis `scripts/clean_data.py` au préalable (nécessite une connexion internet
vers `api.worldbank.org`, sans clé API).

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/indicateurs_cedeao.csv")
df.head()
```

## 1. Aperçu du jeu de données

```python
print(f"Pays : {df['pays'].nunique()}")
print(f"Indicateurs : {df['indicateur'].unique()}")
print(f"Période : {df['annee'].min()} - {df['annee'].max()}")
df.isna().sum()
```

## 2. Croissance du PIB : tendance régionale

```python
croissance = df[df["indicateur"] == "croissance_pib"]
tendance = croissance.groupby("annee")["valeur"].mean()

tendance.plot(marker="o", figsize=(9, 5), title="Croissance du PIB — moyenne CEDEAO")
plt.ylabel("%")
plt.grid(alpha=0.3)
plt.show()
```

## 3. Comparer deux pays

Change les codes ISO3 ci-dessous pour comparer d'autres pays (voir la colonne
`code_iso3` dans le DataFrame).

```python
pays_a, pays_b = "BEN", "GHA"

comp = croissance[croissance["code_iso3"].isin([pays_a, pays_b])]
comp.pivot(index="annee", columns="pays", values="valeur").plot(
    marker="o", figsize=(9, 5), title=f"Croissance du PIB : {pays_a} vs {pays_b}"
)
plt.ylabel("%")
plt.grid(alpha=0.3)
plt.show()
```

## 4. Classement le plus récent, tous indicateurs

```python
derniere_annee = df["annee"].max()
dernier = df[df["annee"] == derniere_annee]

pivot = dernier.pivot(index="pays", columns="indicateur", values="valeur")
pivot.style.background_gradient(cmap="RdYlGn", axis=0)
```
